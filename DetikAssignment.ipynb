```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Kolom yang dimaksud
selected_columns = ['Users', 'Pageviews', 'Pages / Session', 'Avg. Session Duration', 'Avg. Time on Page', 'New Users']

# Menghitung nilai maksimum dan minimum dari setiap kolom
max_values = df[selected_columns].max()
min_values = df[selected_columns].min()

# Menampilkan dalam grafik batang
fig, ax = plt.subplots(figsize=(10, 6))

# Menentukan warna untuk setiap jenis statistik
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# Menentukan lebar batang
bar_width = 0.35
index = np.arange(len(selected_columns))

# Menampilkan batang untuk nilai maksimum
ax.bar(index - bar_width / 2, max_values, bar_width, label='Maksimum', alpha=0.7, color=colors[:len(selected_columns)])

# Menampilkan batang untuk nilai minimum
ax.bar(index + bar_width / 2, min_values, bar_width, label='Minimum', alpha=0.7, color=colors[:len(selected_columns)])

# Menambahkan label di atas batang tanpa miring
for i, (max_value, min_value) in enumerate(zip(max_values, min_values)):
    ax.text(i - bar_width / 2, max_value + 5000, str(max_value), ha='center', va='bottom', color='black', fontsize=10)
    ax.text(i + bar_width / 2, min_value + 5000, str(min_value), ha='center', va='bottom', color='black', fontsize=10)

ax.set_xticks(index)
ax.set_xticklabels(selected_columns)
ax.set_ylabel('Nilai')
ax.set_title('Maksimum dan Minimum Values for Each Column')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Mengurutkan data berdasarkan jumlah Pageviews secara menurun
top_pages = df.sort_values(by='Pageviews', ascending=False).head(19)

# Membuat grafik batang untuk menampilkan halaman atau artikel paling banyak dikunjungi
plt.figure(figsize=(12, 6))
plt.barh(top_pages['Page Title'], top_pages['Pageviews'], color='skyblue')  # Menggunakan barh untuk grafik batang horizontal
plt.xlabel('Pageviews')
plt.ylabel('Page Title')
plt.title('Top 10 Most Visited Pages/Articles')
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

```python
  # Menghitung jumlah setiap jenis konten
content_type_counts = df['contenttype'].value_counts()

# Memilih hanya 5 jenis konten terbanyak
top_5_content_types = content_type_counts.head(5)

# Membuat grafik pie untuk menampilkan distribusi 5 jenis konten terbanyak
fig, ax = plt.subplots(figsize=(12, 6))
colors = plt.cm.Paired.colors  # Menggunakan palet warna yang disediakan oleh matplotlib
explode = (0.1,) * len(top_5_content_types)  # Memberikan efek 'explode' pada semua potongan

# Menambahkan keterangan di samping lingkaran beserta persentase pada keterangan
wedges, texts, autotexts = ax.pie(top_5_content_types, labels=top_5_content_types.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, textprops={'fontsize': 12})
plt.setp(autotexts, size=10)  # Menyetel ukuran teks persentase di dalam potongan

# Menambahkan keterangan di bawah dan kanan dari grafik
legend_labels = [f'{label} ({top_5_content_types[label]} - {autotext.get_text()})' for label, autotext in zip(top_5_content_types.index, autotexts)]
ax.legend(wedges, legend_labels, bbox_to_anchor=(1.1, 0.5), loc="center left", fontsize=12, title="Top 5 Content Types", title_fontsize=14)

# Menampilkan grafik
ax.axis('equal')  # Membuat lingkaran terlihat lebih baik
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

```python
# Menghitung jumlah setiap sumber atau medium
source_medium_counts = df['Source / Medium'].value_counts()

# Memilih hanya 10 sumber atau medium terbanyak
top_10_sources = source_medium_counts.head(10)

# Membuat grafik batang untuk menampilkan distribusi 10 sumber atau medium terbanyak
plt.figure(figsize=(12, 8))
colors = plt.cm.Paired.colors  # Menggunakan palet warna yang disediakan oleh matplotlib

# Menambahkan keterangan di atas batang dengan jumlah pengguna
bars = plt.bar(top_10_sources.index, top_10_sources.values, color=colors)

# Menambahkan keterangan detail di atas setiap batang
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(bar.get_height()), ha='center', va='bottom', fontsize=10)

plt.xlabel('Source / Medium', fontsize=14)
plt.ylabel('Number of Users', fontsize=14)
plt.title('Top 10 Sources / Mediums Bringing Traffic', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

```python
# Menggabungkan data berdasarkan Article ID dan menghitung total Pageviews dan New Users
combined_data = df.groupby('articleid').agg({
    'Page Title': 'first',
    'Pageviews': 'sum',
    'New Users': 'sum'
}).reset_index()

# Mengurutkan data berdasarkan jumlah New Users secara menurun
top_pages = combined_data.sort_values(by='New Users', ascending=False).head(5)

# Membuat subplot
fig, ax = plt.subplots(figsize=(12, 8))

# Grafik batang horizontal untuk menampilkan top pages
bar_height = 0.35
bar_positions = range(len(top_pages['Page Title']))

# Bar untuk Pageviews
bars1 = ax.barh(bar_positions, top_pages['Pageviews'], height=bar_height, color='skyblue', label='Pageviews')

# Bar untuk New Users dengan warna yang berbeda
bars2 = ax.barh([pos + bar_height for pos in bar_positions], top_pages['New Users'], height=bar_height, color='lightcoral', label='New Users')

# Menambahkan keterangan angka di atas setiap batang Pageviews
for bar, value in zip(bars1, top_pages['Pageviews']):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value}', ha='left', va='center', fontsize=10)

# Menambahkan keterangan angka di atas setiap batang New Users
for bar, value in zip(bars2, top_pages['New Users']):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{value}', ha='left', va='center', fontsize=10, color='black')

ax.set_ylabel('Page Title', fontsize=14)
ax.set_xlabel('Counts', fontsize=14)
ax.set_title('Top 5 Most Visited Pages with New Users Information', fontsize=16)
ax.set_yticks([pos + bar_height / 2 for pos in bar_positions])
ax.set_yticklabels([f"{title}\n({article_id})" for title, article_id in zip(top_pages['Page Title'], top_pages['articleid'])], fontsize=12)
ax.legend()

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler



# Memilih fitur untuk prediksi
features = df[['Users', 'Pages / Session', 'Avg. Session Duration', 'Avg. Time on Page', 'New Users']]

# Memilih variabel target
target = df['Pageviews']

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardisasi fitur
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Membuat model regresi linier
model = LinearRegression()

# Melatih model
model.fit(X_train_scaled, y_train)

# Memprediksi data uji
y_pred = model.predict(X_test_scaled)

# Evaluasi model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Visualisasi
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.title('Perbandingan Nilai Sebenarnya dan Nilai Prediksi Pageviews')
plt.xlabel('Pageviews Sebenarnya')
plt.ylabel('Pageviews Prediksi')
plt.show()
```

```python
# Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Pilih fitur (features) dan target
features = df[['Users', 'Pages / Session', 'Avg. Session Duration', 'New Users']]
target_pageviews = df['Pageviews']
target_new_users = df['New Users']

# Bagi dataset menjadi data latih dan data uji untuk Pageviews
X_train_pv, X_test_pv, y_train_pv, y_test_pv = train_test_split(features, target_pageviews, test_size=0.2, random_state=42)

# Buat model Regresi Linear untuk Pageviews
model_pageviews = LinearRegression()
model_pageviews.fit(X_train_pv, y_train_pv)
predictions_pv = model_pageviews.predict(X_test_pv)

# Evaluasi model Pageviews
mae_pv = mean_absolute_error(y_test_pv, predictions_pv)
mse_pv = mean_squared_error(y_test_pv, predictions_pv)
r2_pv = r2_score(y_test_pv, predictions_pv)

print('Pageviews Model:')
print('Mean Absolute Error:', mae_pv)
print('Mean Squared Error:', mse_pv)
print('R-squared:', r2_pv)

# Bagi dataset menjadi data latih dan data uji untuk New Users
X_train_nu, X_test_nu, y_train_nu, y_test_nu = train_test_split(features, target_new_users, test_size=0.2, random_state=42)

# Buat model Regresi Linear untuk New Users
model_new_users = LinearRegression()
model_new_users.fit(X_train_nu, y_train_nu)
predictions_nu = model_new_users.predict(X_test_nu)

# Evaluasi model New Users
mae_nu = mean_absolute_error(y_test_nu, predictions_nu)
mse_nu = mean_squared_error(y_test_nu, predictions_nu)
r2_nu = r2_score(y_test_nu, predictions_nu)

print('\nNew Users Model:')
print('Mean Absolute Error:', mae_nu)
print('Mean Squared Error:', mse_nu)
print('R-squared:', r2_nu)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Visualisasi untuk Pageviews
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)

plt.scatter(y_test_pv, predictions_pv, color='blue', alpha=0.5)
plt.plot([min(y_test_pv), max(y_test_pv)], [min(y_test_pv), max(y_test_pv)], linestyle='--', color='red', linewidth=2)
plt.title('Pageviews: Actual vs. Predicted')
plt.xlabel('Actual Pageviews')
plt.ylabel('Predicted Pageviews')
plt.grid(True)

# Visualisasi untuk New Users
plt.subplot(1, 2, 2)

plt.scatter(y_test_nu, predictions_nu, color='green', alpha=0.5)
plt.plot([min(y_test_nu), max(y_test_nu)], [min(y_test_nu), max(y_test_nu)], linestyle='--', color='orange', linewidth=2)
plt.title('New Users: Actual vs. Predicted')
plt.xlabel('Actual New Users')
plt.ylabel('Predicted New Users')
plt.grid(True)

plt.tight_layout()
plt.show()
```
